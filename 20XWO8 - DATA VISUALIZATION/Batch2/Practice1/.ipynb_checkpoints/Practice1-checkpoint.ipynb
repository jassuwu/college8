```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Practice1.csv')
df
```

```python
plt.bar(df['Age group'], df['No. of bad loans'], color='orange')
plt.bar(df['Age group'], df['No. of good loans'], bottom=df['No. of bad loans'], color='skyblue')
plt.gca().set(ylim=(0, 250))
plt.xticks(rotation=45)
plt.show()
```

```python
df['Bad rate'] = df['No. of bad loans'] / df['Total no. of loans']
plt.plot(df['Age group'], df['Bad rate'])
plt.xticks(rotation=45)
plt.show()
```

# PIE CHART

```python
plt.figure(figsize=(10, 6))
plt.pie(df['Total no. of loans'], labels=df['Age group'], autopct='%1.1f%%')
plt.title('Pie Chart of Total Loans Across Age Groups')
plt.show()
```

# BOX PLOT

```python
plt.figure(figsize=(10, 6))
plt.boxplot([df['Total no. of loans'], df['No. of good loans'], df['No. of bad loans']], labels=['Total Loans', 'Good Loans', 'Bad Loans'])
plt.title('Box Plots of Total Loans, Good Loans and Bad Loans')
plt.show()
```
