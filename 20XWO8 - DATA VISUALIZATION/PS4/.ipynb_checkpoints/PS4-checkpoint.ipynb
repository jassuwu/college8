```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
df
```

```python
# 1
plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
```

```python
# 2
plt.plot(df['month_number'], df['total_units'], ':ro', markersize=3)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
```

```python
for col in df.columns[:-2]:
    plt.plot(df['month_number'], df[col], label=col)
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
```

```python
# 4
plt.scatter(df['month_number'], df['toothpaste'])
```

```python
# 5
df[['month_number', 'facecream', 'facewash']].plot(kind='bar', x='month_number')
```

```python
# 6
df[['month_number', 'bathingsoap']].plot(kind='bar', x='month_number')
```

```python
# 7
plt.hist(df['total_profit'], bins=5, edgecolor='black')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit for Each Month')
plt.show()
```

```python
# 8
plt.pie(df['total_units'], labels=df['month_number'])
```
