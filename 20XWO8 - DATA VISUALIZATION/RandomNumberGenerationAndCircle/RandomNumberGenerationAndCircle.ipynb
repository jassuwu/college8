```python
import collections
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Random Number Generation

```python
fig, axs = plt.subplots(3)

n, bins, patches = axs[0].hist(np.random.normal(0, 0.1, 50), 30)
axs[0].plot(np.convolve(bins, np.ones(2)/2, mode='valid'), n, linestyle='-', color='red', marker='o')

n, bins, patches = axs[1].hist(np.random.uniform(size=50), 30)
axs[1].plot(np.convolve(bins, np.ones(2)/2, mode='valid'), n, linestyle='-', color='red', marker='o')

n, bins, patches = axs[2].hist(np.random.poisson(5, 50), 30)
axs[2].plot(np.convolve(bins, np.ones(2)/2, mode='valid'), n, linestyle='-', color='red', marker='o')

plt.show()
```
